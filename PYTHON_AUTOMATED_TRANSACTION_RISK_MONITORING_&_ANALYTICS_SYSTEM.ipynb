# AUTOMATED TRANSACTION RISK MONITORING & ANALYTICS SYSTEM

```python
import pandas as pd
import os

# ===============================================
# CONFIGURATION
# ===============================================
TRANSACTION_SOURCE_DIRECTORY = r'C:\Users\ARJUN\Downloads\files'
RISK_ENTITY_REFERENCE_PATH = r'C:\Users\ARJUN\Downloads\destination_folder\VPAs.xlsx'


# ===============================================
# TRANSACTION DATA INGESTION
# ===============================================
transaction_dataset_list = []

for source_file in os.listdir(TRANSACTION_SOURCE_DIRECTORY):
    if source_file.endswith('.csv'):
        
        # Build the output path (folder path + file name)
        source_file_path = os.path.join(TRANSACTION_SOURCE_DIRECTORY, source_file)
        
        # Read the files into a dataframe
        transaction_dataset = pd.read_csv(source_file_path)
        
        # Append the files
        transaction_dataset_list.append(transaction_dataset)
        
# Consolidate all transaction datasets
consolidated_dataframes = pd.concat(transaction_dataset_list, ignore_index = True)


# ===============================================
# MERCHANT PERFORMANCE ANALYSIS
# ===============================================
merchant_performance_summary = (
    consolidated_dataframes.groupby('MERCHANT')['AMOUNT']
    .agg(['count','sum'])
    .rename(
        columns = {'count':'Total_Transaction_Count',
                   'sum':'Total_Transaction_Value'})
    .sort_values(by = 'Total_Transaction_Value', ascending = False)
)

merchant_performance_summary['Average_Ticket_Size'] = (
    merchant_performance_summary['Total_Transaction_Value'] 
    / 
    merchant_performance_summary['Total_Transaction_Count']
)


# ===============================================
# TEMPORAL DATA PROCESSING
# ===============================================
consolidated_dataframes['COMPLETION-DATE'] = pd.to_datetime(consolidated_dataframes['COMPLETION-DATE'])
consolidated_dataframes['DATE'] = consolidated_dataframes['COMPLETION-DATE'].dt.date


# ===============================================
# CUSTOMER ACTIVITY ANALYSIS
# ===============================================
customer_activity_summary = (
    consolidated_dataframes.groupby('PAYERVA')
    .agg(Total_Transaction_Count = ('AMOUNT','count'), 
         Total_Transaction_Amount = ('AMOUNT','sum'), 
         Total_Active_Days = ('DATE','nunique'))
    .sort_values(by = 'Total_Active_Days', ascending = False)
)


# ===============================================
# HIGH RISK ENTITY MONITORING
# ===============================================
risk_reference_df = pd.read_excel(RISK_ENTITY_REFERENCE_PATH)

high_risk_entities = risk_reference_df['Payer VPA'].unique()

high_risk_entity_transactions = consolidated_dataframes[consolidated_dataframes['PAYERVA'].isin(high_risk_entities)]


# ===============================================
# REPORTING OUTPUT
# ===============================================
print('MERCHANT PERFORMANCE SUMMARY')
display(merchant_performance_summary.head(10))

print('CUSTOMER ACTIVITY SUMMARY')
display(customer_activity_summary.head(10))

print('HIGH-RISK ENTITY TRANSACTION ACTIVITY')
display(high_risk_entity_transactions.head(10))
```
